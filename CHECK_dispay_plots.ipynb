```python
import glob
import numpy as np
import xarray as xr
from scipy import interpolate
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
import proplot as pplt
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.filterwarnings('ignore')
```

```python
path = '/ocean/projects/atm200007p/sferrett/'
e3sm = np.squeeze(xr.open_dataset(path+'e3sm/E3SM_ANN_000101_000312_climo_PRECL_PRECC_PHIS.nc'))
era5 = xr.merge((xr.open_dataset(path+'era5/ERA5_3D_195906_202108_monthly.nc'),
                 xr.open_dataset(path+'era5/ERA5_2D_195906_202108_monthly.nc')),compat='override')
era5 = era5.reindex(latitude=era5.latitude[::-1]).reindex(level=era5.level[::-1])
era5 = era5.rename({'latitude':'lat','longitude':'lon'})
```

```python
lat1=10; lat2=20; lon1=50; lon2=90; season=['JUN','JUL','AUG','JJA']; years=63
pplt.rc.reso='hi'; colors=np.asarray(['red6','blue6','green6']); lights=np.asarray(['red2','blue1','green2'])
```

```python
def varprofile(vardata):
    lonslice = vardata.lon.sel(lon=slice(lon1,lon2)).to_numpy()
    varslice = vardata.sel(lon=slice(lon1,lon2),lat=slice(lat1,lat2)).mean(dim='lat').to_numpy()
    transform = interpolate.interp1d(lonslice,varslice,bounds_error=False)
    data = xr.DataArray(transform(e3sm.lon.sel(lon=slice(lon1,lon2)))).rename({'dim_0':'time','dim_1':'lon'})
    databymonth = xr.concat((data[0::3],data[1::3],data[2::3]),dim='month')
    return databymonth
```

```python
z   = e3sm.PHIS.sel(lat=slice(lat1,lat2),lon=slice(lon1,lon2)).mean(dim='lat')/9.80665/1000
lon = e3sm.lon.sel(lon=slice(lon1,lon2))
pr  = varprofile(era5.mtpr*86400)
```

```python
fig,gs = pplt.figure(width=9,height=5),pplt.GridSpec(nrows=2,ncols=3,hratios=(0.4,1))
fig.subplot(gs[0,0]); fig.subplot(gs[0,1]); fig.subplot(gs[0,2]); fig.subplot(gs[1,:])
axs = fig.get_axes()
for i,ax in enumerate(axs):
    ox = ax.twinx()
    ox.fill_between(lon,z,fc='grey',alpha=0.2,zorder=0)
    ox.format(ylim=(0,0.8),ylabel='',yticks='null')
    fig.format(suptitle='ERA5 (1959 - 2021) Mean '+str(lat1)+'-'+str(lat2)+'$^\circ$N Cross-Slope Profile',
           abc='A.',abcloc='ul',titleloc='ul',title=season[i],
           xlabel='Longitude [$^\circ$E]',xlim=(50,90),xticks=5,
           ylabel='Precipitation [mm day$^{-1}$]',ylim=(0,35),yticks=5)  
for i,ax in enumerate(axs[0:3]):
    for j in range(years):
        ax.plot(lon,pr[i,j],colors=colors[i],alpha=0.1,linewidth=0.7)
        ax.plot(lon,pr[i].mean(dim='time'),colors=colors[i],linewidth=2)
        axs[3].fill_between(lon,pr[i].min(dim='time'),pr[i].max(dim='time'),fc=lights[i],alpha=0.01)
        axs[3].plot(lon,pr[i].mean(dim='time'),colors=colors[i],linewidth=2)
```
